# Using the Qwen2.5-1.5B Model for Text Generation
---

This notebook shows how to load and use the Qwen/Qwen2.5-1.5B model from Hugging Face to answer a simple question.

Steps:

- Loading the dependencies: Importing the AutoModelForCausalLM and AutoTokenizer classes from the transformers library  https://huggingface.co/docs/transformers/index
- Loading the Model and the Tokenizer: We specify the name of the pre-trained model (Qwen/Qwen2.5-1.5B) and load the model itself together with its associated tokenizer. The tokenizer is essential to convert the text into a format the model can understand.
- Preparing the input: We define the question to ask the model. This question is then passed to the tokenizer, which converts it into input_ids (numeric identifiers for each token/word) and creates an attention_mask.
- Generating the answer: We use the model's generate() method to produce text.
inputs["input_ids"]: The tokens of the question.
attention_mask: Tells the model which tokens are important (not padding).
max_new_tokens=50: Limits the length of the newly generated part (the answer) to 50 tokens.
num_return_sequences=1: We request a single possible answer.
- Decoding and Displaying: The raw model output (a sequence of IDs) is decoded by the tokenizer to convert it back into readable text. We then display the original question and the generated answer.


```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Select the device to use (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 1. Load the model and the tokenizer
model_name = "Qwen/Qwen2.5-1.5B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

# 2. Prepare the question for the model
question = "What is the capital of France?"
inputs = tokenizer(question, return_tensors="pt", padding=True, truncation=True)
inputs = {k: v.to(device) for k, v in inputs.items()}

# 3. Generate an answer with the model
with torch.no_grad():
    output = model.generate(
        inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_new_tokens=50,  # Replaces max_length
        num_return_sequences=1
    )
```

```python
# 4. Decode and display the answer
response = tokenizer.decode(output[0], skip_special_tokens=True)
print("Question:", question)
print("Answer:", response)
```
